```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = None 

glassdoor = pd.read_csv('eda_data.csv')
glassdoor.head()
```

```python
# Remove 12 columns to obtain modified glassdoor (gldr)

gldr = glassdoor.drop(['Job Title','Job Description', 'Founded', 'Type of ownership', 'Industry','hourly',
                       'employer_provided', 'min_salary', 'max_salary', 'company_txt', 'desc_len', 'num_comp'], axis = 1)

gldr.info()

print(gldr.isnull().values.sum())  # Counts the number of Null cells

print(gldr.columns[gldr.isna().any()].tolist())  # https://stackoverflow.com/questions/49590874/dataframe-object-has-no-attribute-isna
```

```python
# Change Column names
gldr.rename({
             'Unnamed: 0': 'Job ID', 'Salary Estimate': 'Salary Range', 'Location': 'Job Location', 'Headquarters': 'Company HQ',
             'avg_salary': 'Annual Salary', 'job_state': 'Job Region', 'python_yn': 'Python Y/N', 'R_yn': 'R Y/N',
             'spark': 'Spark Y/N', 'aws': 'AWS Y/N', 'excel': 'Excel Y/N', 'job_simp': 'Job Type', 'seniority': 'Seniority Y/N/na'
             }, axis = 1, inplace = True)
list(gldr.columns)
```

```python
gldr = gldr.drop(['Job ID','Salary Range', 'Company Name', 'Job Location', 'Company HQ', 'Seniority Y/N/na'], axis = 1)
list(gldr.columns)

gldr.rename({
             'Rating': 'Company_Rating', 'Size': 'Company_Size', 'Annual Salary': 'Annual_Salary', 'Job Region': 'Job_Region', 
             'same_state': 'Same_State', 'age': 'Company_Age', 'Python Y/N': 'Python_YN', 'R Y/N': 'R_YN',
             'Spark Y/N': 'Spark_YN', 'AWS Y/N': 'AWS_YN', 'Excel Y/N': 'Excel_YN', 'Job Type': 'Job_Type'
             }, axis = 1, inplace = True)
list(gldr.columns)
```

```python
# Verifying if a company has competitors.
# Under 'competitors', if there is no competing company, '-1' is used as filler,
# if there is competition, names of competing companies are listed.
# The following code modifies the cell-content under 'Competitors' as 'yes' and 'no',
# for other competing company and no competition, respectively.

gldr.loc[gldr['Competitors']!='-1', 'Competitors'] = 'yes'  
gldr.loc[gldr['Competitors']=='-1', 'Competitors'] = 'no'

print(gldr['Competitors'].head(10))
# gldr.info()
```

```python
# Changing job_state to job_region; bar charts 

# 5 Regions according to National Geographic

West = "WA-OR-ID-MT-CA-NV-UT-WY-CO-AK-HI"  # 11 States
Midwest = "ND-SD-NE-KS-MN-IA-MO-WI-IL-MI-IN-OH"    # 12 States
Northeast = "NY-VT-ME-NH-MA-RI-CT-NJ-PA"  # 9 States
Southeast = "WV-DE-MD-VA-KY-TN-NC-AR-LA-MS-AL-GA-SC-FL-DC"  # 14 States
Southwest = "AZ-NM-TX-OK"    # 4 States

for i in range(0, len(gldr)):

  if West.find(gldr.iloc[i, gldr.columns.get_loc('Job_Region')])!=-1:
    (gldr.iloc[i,  gldr.columns.get_loc('Job_Region')]) = "West"

  elif Midwest.find(gldr.iloc[i, gldr.columns.get_loc('Job_Region')])!=-1:
    (gldr.iloc[i,  gldr.columns.get_loc('Job_Region')])= "Midwest"

  elif Northeast.find(gldr.iloc[i, gldr.columns.get_loc('Job_Region')])!=-1:
    (gldr.iloc[i,  gldr.columns.get_loc('Job_Region')])= "Northeast"

  elif Southeast.find(gldr.iloc[i, gldr.columns.get_loc('Job_Region')])!=-1:
    (gldr.iloc[i,  gldr.columns.get_loc('Job_Region')])= "Southeast"

  else:
    (gldr.iloc[i, gldr.columns.get_loc('Job_Region')])= "Southwest"

gldr['Job_Region'].tail(10)
```

```python
# Counts of categories in some features 

gldr['Competitors'] = gldr['Competitors'].astype('object')          
print('Does the Company have competitor(s)? yes/no\n', gldr['Competitors'].value_counts(), '\n')
print('Percentage of each catagory:\n', gldr['Competitors'].value_counts(normalize=True).mul(100).round(1).astype(str), '%\n')
print('-----------------------------------------------\n\n')

gldr.loc[gldr['Same_State']== 1, 'Same_State'] = 'yes'
gldr.loc[gldr['Same_State']== 0, 'Same_State'] = 'no'       
print('Is the job in the same state as the company headquarters? yes/no\n', gldr['Same_State'].value_counts(), '\n')
print('Percentage of each catagory:\n', gldr['Same_State'].value_counts(normalize=True).mul(100).round(1).astype(str), '%\n')
print('-----------------------------------------------\n\n')

lang_list = ['Python_YN', 'R_YN', 'Spark_YN', 'AWS_YN', 'Excel_YN']

def language(lang):
  gldr.loc[gldr[lang]== 1, lang] = 'yes'
  gldr.loc[gldr[lang]== 0, lang] = 'no'          
  print(f'Does the job require knowledge of {lang}?\n', gldr[lang].value_counts(), '\n')
  print('Percentage of each catagory:\n', gldr[lang].value_counts(normalize=True).mul(100).round(1).astype(str), '%\n')
  print('-----------------------------------------------\n\n')

for i in lang_list:
  language(i)
```

```python
# Deleting records with unknown values (-1) in the attribute 'Rating'; number of records areduced from 742 to 731
gldr.loc[gldr['Company_Rating'] ==-1, 'Company_Rating'] = ''
gldr['Company_Rating'].replace('', np.nan, inplace=True)
gldr.dropna(inplace = True)
#gldr['Company_Rating'].tail(50)
gldr.info()
```

```python
# Deleting records with unknown values (-1) in the attribute 'Rating'; number of records areduced from 731 to 730
gldr.loc[gldr['Sector'] == '-1', 'Sector'] = ''  # Here -1 implies data is unknown
gldr['Sector'].replace('', np.nan, inplace=True)
gldr.dropna(inplace = True)  #  Drop records with missing data
gldr.info()
# Number of records is reduced from 731 to 730
```

```python
# 'Company_Age' is the number of years since a company was established to the year 2020
# -1 is filler for Unknown age of a company
# Drop records with Unknown age of a company 
# No. of records reduced 730 to 690

gldr.loc[gldr['Company_Age'] == -1, 'Company_Age'] = ''
gldr['Company_Age'].replace('', np.nan, inplace=True)
gldr.dropna(inplace = True)
gldr['Company_Age'] = gldr['Company_Age'].astype('float64')
gldr.info()
```

```python
#Modify Company_Size column to a float
#strip the string "employees" from the column "Company_Size"
gldr['Company_Size'] = gldr['Company_Size'].str.strip('employees')

#remove the string " to " from the column "Company_Size"
gldr['Company_Size'] = gldr['Company_Size'].str.replace(' to ', ' ')
gldr['Company_Size'] = gldr['Company_Size'].str.replace('+', '')

#if there is a space in the column "Company_Size", split the string and cast the values to floats, then add those values together and divide by 2
gldr['Company_Size'] = gldr['Company_Size'].apply(lambda x: sum([float(i) for i in x.split()])/2 if ' ' in x else x)

gldr.head()
```

```python
gldr.iloc[1, 12] = 'yes'  # Error correction 'Same_State', yes/no
```

```python
gldr.reset_index(drop = True, inplace = True)
print(gldr.shape[0])
gldr.to_csv('gldr.csv', index = False)
```

```python
#Box plots displaying outliers for quantitative attributes
gldr1 = gldr.select_dtypes(exclude = ['object'])
gldr1 = gldr1.select_dtypes(exclude = ['category'])
gldr1.info()

for column in gldr1:
        plt.figure(figsize=(17,1))
        sns.boxplot(data = gldr1, x = column)
```

```python
# Box & Whiskers plot and Outlier Numbers for features: 'Company_Rating', 'Annual_Salary', 'Company_Age', 'Company_Size'

c_list = ['deepskyblue', 'orchid', 'limegreen', 'gold']
feature_nm = ['Company_Rating', 'Annual_Salary', 'Company_Age', 'Company_Size']

def box_limits(name, clr):

  plt.figure(figsize=(15,5))
  sns.boxplot(data = gldr[[name]], color = clr, orient = 'h')
  plt.show()

  q1 = gldr[name].quantile(.25)
  q3 = gldr[name].quantile(.75)
  iqr = q3 - q1

  mean = gldr[name].mean()
  median = gldr[name].median()
  stand_dv = gldr[name].std()

  print(f'Quartiles and Limits for company {name}\n')
  print(f'First Quartile: {q1:.2f}, Third Quartile: {q3:.2f}, IQ: {iqr:.2f}, Mean: {mean:.2f}, Median: {median:.2f}, Standard Deviation: {stand_dv: .2f}\n')

  li = q1 - 1.5*(iqr)
  ui = q3 + 1.5*(iqr)
  print(f'Inner Lower Limit: {li:.2f}, Inner Upper Limit: {ui:.2f}\n')

  lo = q1 - 3*(iqr)
  uo = q3 + 3*(iqr)
  print(f'Outer Lower Limit: {lo:.2f}, Outer Upper Limit: {uo:.2f}\n')

  print(f'\nMild Outliers:\n{gldr[(gldr[name] < li)].count() + gldr[(gldr[name] > ui)].count()}\n')
  print('Records of Mild Ouliers:\n')
  print('\n', gldr[(gldr[name] < li)] , '\n',  gldr[(gldr[name] > ui)], '\n')

  print(f'\nExtreme Outliers:\n{gldr[(gldr[name] < lo)].count() + gldr[(gldr[name] > uo)].count()}\n')
  print('Records of Extreme Ouliers:\n')
  print('\n', gldr[(gldr[name] < lo)] , '\n',  gldr[(gldr[name] > uo)], '\n')


for j in range(len(feature_nm)):
  box_limits(feature_nm[j], c_list[j])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
money = pd.read_csv("gldr.csv")
money.head()
```

```python
money.info()
```

**********************************************************************************************************************
<h4>Note:</h4> 
<P>It should be noted that decision-tree machine learning requires all data to be in numerical format. Therefore, all columns of "object" data type are converted to "category" data type. The category data type is then converted to numerical data type using the cat.codes method. This method assigns a numerical value to each category in a column. The numerical value is assigned to the region based on its location clockwise in the continental United States with the "Northeast" represented as 1. This is done exclusively for the 'Job_Region' column. The data is then ready for machine learning.</p>

**********************************************************************************************************************

```python
money.describe()
```

```python
money.rename(columns={'Excel_YN': 'Excel', 'Python_YN': 'Python', 'R_YN': 'R', 'Spark_YN': 'Spark', 'AWS_YN': 'AWS'}, inplace=True)

money.head()
```

```python
#replace region codes with integers in region column
#West = 4 or "WA-OR-ID-MT-CA-NV-UT-WY-CO-AK-HI"  # 11 States
#Midwest = 5 or "ND-SD-NE-KS-MN-IA-MO-WI-IL-MI-IN-OH"    # 12 States
#Northeast = 1 or "NY-VT-ME-NH-MA-RI-CT-NJ-PA"  # 9 States
#Southeast = 2 or "WV-DE-MD-VA-KY-TN-NC-AR-LA-MS-AL-GA-SC-FL-DC"  # 14 States
#Southwest = 3 or "AZ-NM-TX-OK"    # 4 States

money['Job_Region'] = money['Job_Region'].replace(['West'], 4)
money['Job_Region'] = money['Job_Region'].replace(['Midwest'], 5)
money['Job_Region'] = money['Job_Region'].replace(['Northeast'], 1)
money['Job_Region'] = money['Job_Region'].replace(['Southeast'], 2)
money['Job_Region'] = money['Job_Region'].replace(['Southwest'], 3)

money.head()

```

```python
money.drop(['Revenue', 'Job_Type', 'Sector'], axis=1, inplace=True)

money.head()
```

```python
money['Python'] = money['Python'].map({'yes': 1, 'no': 0})
money['R'] = money['R'].map({'yes': 1, 'no': 0})
money['Spark'] = money['Spark'].map({'yes': 1, 'no': 0})
money['AWS'] = money['AWS'].map({'yes': 1, 'no': 0})
money['Excel'] = money['Excel'].map({'yes': 1, 'no': 0})
money['Competitors'] = money['Competitors'].map({'yes': 1, 'no': 0})
money['Same_State'] = money['Same_State'].map({'yes': 1, 'no': 0})

money.head()
```

```python
money['Above_Median'] = np.where(money['Annual_Salary']>96.0, 1, 0)

money.head()
```

<h1>Data Modeling</h1>

**********************************************************************************************************************

<p>Due to the high proportion of binary features, a decision tree classifier is used to determine the most important features.</p>

**********************************************************************************************************************

```python
#split data into training and testing sets
feature_set = ['Company_Rating', 'Company_Size', 'Job_Region', 'Competitors', 'Same_State', 'Company_Age', 'Python', 'R', 'Spark', 'AWS', 'Excel']
features = money[feature_set]
target = money['Above_Median']

X_train, X_test, y_train, y_test =\
      train_test_split(features, target, test_size=0.3, random_state=1)
```

```python
model = DecisionTreeClassifier(max_depth=5, random_state=1)
model.fit(X_train, y_train)

#predict the test set
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix

cm  = confusion_matrix(y_test, y_pred)
ax =sns.heatmap(cm, annot=True, fmt="d", cmap="plasma")
ax.set(xlabel="Predicted", ylabel="Actual")
plt.show()
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print()

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
